```python
from ollama import chat
from datetime import datetime

start = datetime.now()
response = chat(
    model='thiagomoraes/medgemma-4b-it:Q4_K_S',
    messages=[{'role': 'user', 'content': 'What are the most common causes of persistent chest pain in adults, and how can doctors differentiate between cardiac and non-cardiac origins'}],
)
print(response.message.content)
end = datetime.now()
```

```python
print(f"Total time: {end - start}")
```
