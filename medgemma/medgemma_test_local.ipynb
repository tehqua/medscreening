```python
from transformers import pipeline
import torch
from IPython.display import display, Markdown
from datetime import datetime
```

```python
# Model path
downloaded_path = "./google_medgemma_4b_it"
start = datetime.now()
# Initialize pipeline
pipe = pipeline(
    "text-generation",
    model=downloaded_path,
    device="cuda",
)

# Prompt
prompt = "What are the most common causes of persistent chest pain in adults, and how can doctors differentiate between cardiac and non-cardiac origins"
role_instruction = "You are a helpful medical assistant."
system_instruction = role_instruction

max_new_tokens = 300

# Message structure
messages = [
    {
        "role": "system",
        "content": [
            {
                "type": "text",
                "text": system_instruction
            }
        ],
    },
    {
        "role": "user",
        "content": [
            {
                "type": "text",
                "text": prompt
            }
        ],
    },
]

# Generate output
output = pipe(
    messages,
    max_new_tokens=max_new_tokens,
    do_sample=False
)

response = output[0]["generated_text"][-1]["content"]
end = datetime.now()
```

```python
print(f"Total time: {end - start}")
```

```python
display(Markdown(f"---\n\n**[ User ]**\n\n{prompt}\n\n---"))
display(Markdown(f"**[ MedGemma ]**\n\n{response}\n\n---"))
```
